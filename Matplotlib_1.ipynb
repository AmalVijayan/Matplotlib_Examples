```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("sample_data.csv")
```

```python
data.head(10)
```

```python
plt.plot(data['A'], c = 'b')
plt.title("Change In A")
plt.xlabel("Index")
plt.ylabel("A")
plt.legend()
```

```python
plt.scatter(x = data['A'],y = data['B'],s = 500, c = '#3FD977', marker = "*",alpha = 1, linewidths = 1, edgecolors='b')
plt.title("Scattered A vs B")
plt.xlabel("A")
plt.ylabel("B")
```

```python
#Initializing an empty figure object
fig = plt.figure(tight_layout=True)  

#Initializing a 2x2 grid of Axes inside the figure object
fig, axes = plt.subplots(2, 2)  

# Setting a title
fig.suptitle('Change In A,B & C ') 

#Plotting Index vs A in subplot[0,0]-- 0th row and 0th column
axes[0,0].plot(data['A'])

#Plotting Index vs B in subplot[0,1]-- 0th row and 1st column
axes[0,1].plot(data['B'])

#Plotting Index vs C in subplot[1,0]-- !st row and 0th column
axes[1,0].plot(data['C'])

#Plotting D vs A for A values greater than 0 in subplot[1,1]
axes[1,1].bar(data[data['A']>0]['D'], data[data['A']>0]['A'])

```

```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec 

fig = plt.figure(tight_layout=True)


#specifies the geometry of the grid that a 
#subplot will be placed. The number of rows and number of columns of the grid need to be set.
gs = gridspec.GridSpec(2, 2)  


#Subplot 1
ax = fig.add_subplot(gs[0, :])
ax.scatter(x = data['A'], y = data['B'], c='r')
ax.set_ylabel('B')
ax.set_xlabel('A')


#Subplot 2
ax = fig.add_subplot(gs[1, 0])
ax.plot(data['A'])
ax.set_ylabel('A')
ax.set_xlabel('Index')
fig.align_labels()

#Subplot 3
ax = fig.add_subplot(gs[1, 1])
ax.plot(data['B'])
ax.set_ylabel('B')
ax.set_xlabel('Index')
fig.align_labels()    

```
